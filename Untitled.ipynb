```python
import surprise
import numpy
import pandas as pd
```

```python
from surprise import Dataset, Reader

stUserjev = 50 #stevilo koliko userjev bomo ocenjevali
stFilmov = 50 #stevilo koliko filmov bomo ocenjevali

raw = pd.read_csv('ratings.csv') #preberi csv
raw = raw[["userId","movieId","rating"]] #izloƒçi nepotrebne stolpce

reader = Reader(rating_scale=(1,5)) #nastavi skalo od 1 do 5
data = Dataset.load_from_df(raw, reader) 
```

```python
from surprise.model_selection import train_test_split

#nastavi trainset-testset 0.8-0.2
trainset, testset = train_test_split(data,test_size=0.2)
```

```python
from surprise import KNNBasic

#uporabi KNN algoritem
predictor = KNNBasic()

#treniraj algoritem na trainsetu
predictor.fit(trainset)
```

```python
#pivotiraj tabelo, NA->0
user_movie = raw.pivot_table(index='userId',columns='movieId',values= 'rating')
user_movie = user_movie.fillna(0)
user_movie_knn = user_movie
user_movie_algo = user_movie
user_movie_naive = user_movie
```

```python
#prikazi original tabelo kako izgleda
user_movie.head(10)
```

```python
#Vstavi predvidene ocene ki s pomoƒçjo KNN algoritma za vsakega userja

for i in range(stUserjev):  #usersLen
    for j in range(stFilmov): #moviesLen
        sample = user_movie_knn.sample(1)
        user_id = i
        book_id = j # id od enega sampla
        

        p = predictor.predict(uid=user_id, iid=book_id)
        predicted_rating = p.est
        predicted_rating = str(round(predicted_rating, 2))

        user_movie_knn.iloc[i][j] = predicted_rating
```

```python
user_movie_knn.head(10)
```

```python
#predvidi ocene s pomocjo KNN
predictions = predictor.test(testset) 
```

```python
from surprise import accuracy

#izracunaj MAE
mae_KNN = accuracy.mae(predictions, verbose=True)
```

```python
#Slope one 

from surprise import SlopeOne
algo = SlopeOne()

#treniraj algoritem na trainsetu
algo.fit(trainset)

#predvidi ocene s pomocjo SlopeOne
predictions2 = algo.test(testset)
```

```python
from surprise import accuracy

#izracunaj MAE
mae_slope = accuracy.mae(predictions2, verbose=True)
```

```python
#Izracunaj MAE za naive predictor

rows_count = len(naive_pred.index)
summ = 0

for i in range(rows_count):
    summ += abs(naive_pred.rating[i] - naive_pred.iloc[i,3])
  
error = summ/rows_count
  
# display
print("MAE naive : ", str(error), "Mae user based: ",mae_KNN, " Mae slope one: ", mae_slope)
```
